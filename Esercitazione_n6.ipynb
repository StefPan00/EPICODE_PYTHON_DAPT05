```python
import pandas as pd
import numpy as np

years = 5 
a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years),                
                            "Anno": np.repeat(list(range(years)), 12),                
                            "Guadagni": np.random.randint(800, 5000, 12*years)})
a
```

```python
#Calcola la somma cumulativa delle vendite per ogni mese utilizzando il metodo .cumsum()
a["CumSum"]=a.Guadagni.cumsum()
```

```python
a
```

```python
a.CumSum.plot()
```

```python
#come sopra ma diviso per anno
a["Years CumSum"]= a.groupby("Anno").Guadagni.cumsum()
a
```

# libreria os

```python
import os 
import pandas as pd

directory= "C:/Users/PANTANO/Desktop/Data Analyst/LEZIONI/3 Terzo Modulo Python/dataset"
lista_db=os.listdir(directory)
```

```python
file_NaN=[]

for file_name in lista_db:
    file_path=os.path.join(directory,file_name)
    try:
        df = pd.read_csv(file_path, encoding="latin1", sep=None, engine='python')
        null_count= df.isna().sum().sum()
        if null_count > 0:
            file_NaN.append(file_name)
            print(file_name)
    except Exception as e:
        print(f"Errore lettura file {file_name}: {e}")

print("File con valori nulli:", file_NaN)
```
