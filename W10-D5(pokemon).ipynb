```python
#• È verosimile che la prima colonna dovrebbe essere un indice?
print(df['#'].dtypes)
print("\nPrima colonna:")
print(df.iloc[:, 0].head())
```

```python
#Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano?
# non combacia perchè python conta da 0
is_index_equal = df.iloc[:, 0].equals(pd.Series(df.index))
print("\nLa prima colonna combacia con l'indice automatico di Pandas?", is_index_equal)

if not is_index_equal:
    df.set_index(df.columns[0], inplace=True)
    print("\nLa prima colonna è stata impostata come indice.")

# 5. Confermare l'indice
print("\nIndice del DataFrame:")
print(df.index)
```
