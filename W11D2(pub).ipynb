```python
#quante attività ci sono nel quadrato di longitudine 9-10 e latitudine 45-46?
lon= (df["Longitudine"]>= 9) & (df["Longitudine"]<=10)
lat= (df["Latitudine"]>= 45) & (df["Latitudine"]<=46)
attività= df[lon & lat]
print("Il numero di attività nel quadrato di longitudine 9-10 e latitudine 45-46 sono:", len(attività))
```

```python
#quante sono le attività nella provincia di Vicenza?
filtro= df["Provincia"]== "VICENZA"
V=df[filtro]
print("le attività nella provicinai di Vicenza sono:", len(V), "attività")
```

```python
#quante enoteche ci sono, e come si chiamano?
enoteca= df["Nome"].str.contains("enoteca", case=False, na=False) #ca sta per case sensitive, metto faslso, e na decido come trattare il resto dei dati e dico di ignorare
C=df[enoteca]
C
#alternativa ciclo for e poi operatore in. 
```

```python
print("enoteche:", len(C))
```

```python
# alternativa 
enoteche=[]

for nome in df["Nome"]:
    nome = str(nome) #casting, trasformiamo in stringa
    if "enoteca" in nome.lower():
        enoteche.append(nome)
        print(nome)

print("\n","nunero di enoteche", len(enoteche))
```

```python
#quante attività complessive ci sono nel Lazio e in Abruzzo?
Lazio= df["Regione"]== "Lazio"
Abruzzo=df["Regione"]== "Abruzzo"
LA= df[Lazio | Abruzzo]
LA
```

```python
import pandas as pd
employees_df = pd.DataFrame({ 'employee_id': [101, 102, 103, 104, 105], 'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 'department_id': [1, 2, 1, 2, 3] })
departments_df = pd.DataFrame({ 'department_id': [1, 2, 3], 'department_name': ['HR', 'IT', 'Finance'], 'location': ['New York', 'San Francisco', 'Chicago'] })
```

```python
employees_df
```

```python
departments_df
```

```python
# Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni sia sui dipendenti 
#che sui dipartimenti, usando la funzione .merge()
pd.merge(left=employees_df, right=departments_df, on="department_id")

#merge lavora come SQL join.
```

```python
#Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente; 
#poi unire i due dataset mediante il metodo .join(). Su pandas join lavora con gli indici.

empolyees_df=employees_df.set_index("department_id")
departmenets_df=departments_df.set_index("department_id")
```

```python
empolyees_df
```

```python
departmenets_df
```

```python
employees_df.join(departmenets_df)
```
